```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import pickle
```

```python
with open('Models/XGB-Model.pkl', 'rb') as file:
    model = pickle.load(file)
```

```python
with open('Models/encoder_media_type.pkl', 'rb') as f:
        encoder_media_type = pickle.load(f)
with open('Models/encoder_traffic_source.pkl', 'rb') as f:
        encoder_traffic_source = pickle.load(f)
with open('Models/encoder_categarious_content.pkl', 'rb') as f:
        encoder_content_category = pickle.load(f)
with open('Models/scaler_numerical.pkl', 'rb') as f:
        scaler = pickle.load(f)
```

```python
sample_post = pd.DataFrame([{
    'media_type': 'Reel',
    'likes': 50000,
    'comments': 2500,
    'shares': 1200,
    'saves': 3500,
    'reach': 150000,
    'impressions': 500000,
    'caption_length': 1500,
    'hashtags_count': 15,
    'followers_gained': 5000,
    'traffic_source': 'Reels Feed',
    'content_category': 'Fashion'
}])
```

```python
encoder_array = encoder_media_type.transform(sample_post[['media_type']])
encoder_df_media_type = pd.DataFrame(
    encoder_array,
    columns=encoder_media_type.get_feature_names_out(['media_type']),
    index=sample_post.index
)
sample_post=pd.concat([sample_post,encoder_df_media_type],axis=1)
sample_post=sample_post.drop(columns='media_type')
```

```python
encoder_array = encoder_traffic_source.transform(sample_post[['traffic_source']])
encoder_df_traffic_source = pd.DataFrame(
    encoder_array,
    columns=encoder_traffic_source.get_feature_names_out(['traffic_source']),
    index=sample_post.index
)
sample_post=pd.concat([sample_post,encoder_df_traffic_source],axis=1)
sample_post=sample_post.drop(columns='traffic_source')
```

```python
sample_post
```

```python
encoder_array = encoder_content_category.transform(sample_post[['content_category']])
encoder_df_content_category = pd.DataFrame(
    encoder_array,
    columns=encoder_content_category.get_feature_names_out(['content_category']),
    index=sample_post.index
)
sample_post=pd.concat([sample_post,encoder_df_content_category],axis=1)
sample_post=sample_post.drop(columns='content_category')
```

```python
numerical_cols = [
    'likes', 'comments', 'shares', 'saves', 'reach',
    'impressions', 'caption_length', 'hashtags_count', 'followers_gained'
]
```

```python
sample_post[numerical_cols]=scaler.transform(sample_post[numerical_cols])
```

```python
prediction=model.predict(sample_post)
```

```python
print(f'The engagment you can get from this is predicted as {prediction}%')
```
