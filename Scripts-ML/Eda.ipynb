```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import pickle
```

```python
df=pd.read_csv('Data.csv')
df.head()
```

```python
df=df.drop(columns=['post_id','upload_date'])
```

```python
df['traffic_source'].unique()
```

```python
Y=df['engagement_rate']
X=df.drop(columns='engagement_rate')
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder_media_type=OneHotEncoder(sparse_output=False,dtype=int)
encoder_traffic_source=OneHotEncoder(sparse_output=False,dtype=int)
encoder_categarious_content=OneHotEncoder(sparse_output=False,dtype=int)
```

```python
numerical_cols = [
    'likes', 'comments', 'shares', 'saves', 'reach',
    'impressions', 'caption_length', 'hashtags_count', 'followers_gained'
]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.2,random_state=42)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
Y_train.head()
```

```python
encoder_array_media_type=encoder_media_type.fit_transform(X_train[['media_type']])
encoder_df_media_type=pd.DataFrame(encoder_array_media_type,columns=encoder_media_type.get_feature_names_out(['media_type']),index=X_train.index)
X_train=pd.concat([X_train,encoder_df_media_type],axis=1)
X_train.head()
```

```python
encoder_array_media_type = encoder_media_type.transform(X_test[['media_type']])

encoder_df_media_type = pd.DataFrame(
    encoder_array_media_type, 
    columns=encoder_media_type.get_feature_names_out(['media_type']), 
    index=X_test.index
)

X_test = pd.concat([X_test, encoder_df_media_type], axis=1)
X_test.head()
```

```python
X_test['media_type_Carousel'].unique()
```

```python
traffic_array_train = encoder_traffic_source.fit_transform(X_train[['traffic_source']])
traffic_df_train = pd.DataFrame(
    traffic_array_train,
    columns=encoder_traffic_source.get_feature_names_out(['traffic_source']),
    index=X_train.index  # Fix Index
)
X_train = pd.concat([X_train, traffic_df_train], axis=1)
```

```python
traffic_array_test = encoder_traffic_source.transform(X_test[['traffic_source']])
traffic_df_test = pd.DataFrame(
    traffic_array_test,
    columns=encoder_traffic_source.get_feature_names_out(['traffic_source']),
    index=X_test.index  # Fix Index
)
X_test = pd.concat([X_test, traffic_df_test], axis=1)
```

```python
content_array_train = encoder_categarious_content.fit_transform(X_train[['content_category']])
content_df_train = pd.DataFrame(
    content_array_train,
    columns=encoder_categarious_content.get_feature_names_out(['content_category']),
    index=X_train.index  # Fix Index
)
X_train = pd.concat([X_train, content_df_train], axis=1)
```

```python
content_array_test = encoder_categarious_content.transform(X_test[['content_category']])
content_df_test = pd.DataFrame(
    content_array_test,
    columns=encoder_categarious_content.get_feature_names_out(['content_category']),
    index=X_test.index  # Fix Index
)
X_test = pd.concat([X_test, content_df_test], axis=1)
```

```python
X_train=X_train.drop(columns=['media_type','content_category','traffic_source'])
X_test=X_test.drop(columns=['media_type','content_category','traffic_source'])
```

```python
X_train.head()
```

```python
X_train.loc[:, numerical_cols] = scaler.fit_transform(X_train[numerical_cols])
X_test.loc[:, numerical_cols] = scaler.transform(X_test[numerical_cols])
```

```python
X_test.head()
```

```python
with open('encoder_media_type.pkl', 'wb') as f:
    pickle.dump(encoder_media_type,f)
with open('encoder_traffic_source.pkl', 'wb') as f:
    pickle.dump(encoder_traffic_source,f)
with open('encoder_categarious_content.pkl', 'wb') as f:
    pickle.dump(encoder_categarious_content,f)
with open('scaler_numerical.pkl', 'wb') as f:
    pickle.dump(scaler,f)
```

```python
os.makedirs('Splitted-Data',exist_ok=True)

X_train.to_csv(path_or_buf='Splitted-Data/X_train_data.csv',index=False)
X_test.to_csv(path_or_buf='Splitted-Data/X_test.csv', index=False)
Y_train.to_csv(path_or_buf='Splitted-Data/Y_train.csv', index=False)
Y_test.to_csv(path_or_buf='Splitted-Data/Y_test.csv', index=False)
```
